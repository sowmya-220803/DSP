```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("train.csv")
data.head()
```

```python
print(data.shape)
print(data.columns)
```

```python
print(data.isnull().sum())
```

```python
corr=data.corr(numeric_only=True)
plt.figure(figsize=(10,8))
sns.heatmap(corr)
plt.show()
```

```python
data=data[['GrLivArea','OverallQual','GarageArea','YearBuilt','SalePrice']]
```

```python
x=data['GrLivArea']
y=data['SalePrice']
m,b=np.polyfit(x,y,1)
plt.scatter(x,y)
plt.plot(x,m*x+b)
plt.xlabel("Living Area")
plt.ylabel("Price")
plt.show()
```
