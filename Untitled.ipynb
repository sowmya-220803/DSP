```python
X = data['GrLivArea'].values
y = data['SalePrice'].values
```

```python
#normalizig the values
X=X/X.max()
y=y/y.max()
```

```python
m=0
b=0
learning_rate=0.1
iterations=1000
n=len(X)
```

```python
for i in range(iterations):
    y_pred=m*X+b
    error=y_pred-y
    
    dm = (2/n) * np.sum(error * X)
    db = (2/n) * np.sum(error)
    
    m = m - learning_rate * dm
    b = b - learning_rate * db
print("slope",m)
print("intercept",b)
    
```

```python
plt.scatter(X,y)
y_pred=m*X+b
plt.plot(X,y_pred,color='red')
plt.show()
```

```python
cost_history = []

for i in range(iterations):

    y_pred = m*X + b
    cost = np.mean((y_pred-y)**2)

    cost_history.append(cost)
```
